##  4. Model Building 6

##  Split data into X and Y

```python
# Arrange data into independent variables and dependent variables
X=df.drop(columns='Class')
y=df['Class'] #target
```

```python
X.describe()
```

##  Scaling Independent Variables

```python
X_Scaled=X.apply(zscore)
```

```python
X_Scaled.describe().T
```

 We have scaled independent variables to corresponding z-score.

We can see Mean becomes close to zero and Standard Deviation becomes 1

##  Split data into train and test with 80:20 proportion

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=10)
```

## Train a Supervised Learning Classification base model using KNN classifier

```python
KNN = KNeighborsClassifier(n_neighbors= 5 , metric = 'euclidean' ) #Building knn with 5 neighbors
```

```python
KNN.fit(X_train, y_train)
predicted_labels = KNN.predict(X_test)
```

```python
print('Accuracy on Training data:',KNN.score(X_train, y_train) )
print('Accuracy on Test data:',KNN.score(X_test, y_test) )
```

Training Acuracy is 0.91 and Testing Accuracy is0.70. Performance is less in test data.

This is due to overfitting of data

## Print all the possible classification metrics for both train and test data

```python
print("classification  Matrix:\n",classification_report(y_test,predicted_labels))
```

##  . Performance Improvement: 4


## Tune the parameters/hyperparameters to improve the performance of the base mode

```python
train_score=[]
test_score=[]
for k in range(1,51):
    KNN = KNeighborsClassifier(n_neighbors= k , metric = 'euclidean' ) 
    KNN.fit(X_train, y_train)
    train_score.append(KNN.score(X_train, y_train))
    test_score.append(KNN.score(X_test, y_test))
```

```python
plt.plot(range(1,51),train_score)
plt.show()
```

Here training accuracy decreases when increase k value

```python
plt.plot(range(1,51),test_score)
plt.show()
```

The maximum accuracy occures when k is less than 15.

We will fix k value as less than 15.

## Clearly showcase improvement in performance achieved

```python
k=[1,3,5,7,9,11,12,13,15]
for i in k:
    KNN = KNeighborsClassifier(n_neighbors=i, metric = 'euclidean' ) #Building knn with 5 neighbors
    KNN.fit(X_train, y_train)
    predicted_labels = KNN.predict(X_test)
    print('Accuracy on Training data for k {} is {}:'.format(i,KNN.score(X_train, y_train)))
    print('Accuracy on Test data for k {} is {}:'.format(i,KNN.score(X_test, y_test)))
    print("classification  Matrix:\n",classification_report(y_test,predicted_labels))
```

For K=12 we have balanced train and test error

we can use k value as 12 because when we increase this value the precision becomes100% for type s

Accuracy has incresed from 70% to 80%

Precision has incresed from 53% to 63% of Normal

Precision has incresed from 45% to 69% of Type_H
